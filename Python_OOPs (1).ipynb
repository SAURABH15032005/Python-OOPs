# **1.What is Object-Oriented Programming (OOP) ?**

-  Object-Oriented Programming (OOP) is a programming technique in which programs are developed using objects.
It focuses on combining data and related functions into a single unit called a class, which makes programs easier to understand, manage, and reuse.

# **2. What is a class in OOP ?**

-  A class in OOP is a blueprint or template used to create objects.
It defines the data members (variables) and member functions (methods) that describe the behavior of objects.

**Example:**

- If Student is a class, then name and roll_no are data members, and study() is a method.
An actual student created from this class is called an object.

# **3.What is an object in OOP ?**

-  An object in OOP is an instance of a class that represents a real-world entity.
It contains actual values for the data members and can use the methods defined in the class.

**Example:**

- If Student is a class, then Student s1 with name = “Rahul” and roll_no = 10 is an object of that class.


# **4.What is the difference between abstraction and encapsulation?**

-  Abstraction focuses on showing only the essential features of an object and hiding the internal implementation.
It is used to reduce complexity and show what an object does.

- Encapsulation focuses on wrapping data and methods together into a single unit (class).
It is used to protect data and controls how data is accessed.

**Example:**

- Abstraction: Using a car—you use steering and brakes without knowing the engine details.

- Encapsulation: The engine parts are enclosed inside the car and cannot be accessed directly.

# **10 What are class and static methods in Python ?**


 **1. Class Methods**


- A class method is a method that works with the class itself, rather than an instance (object).
It can access or modify class-level attributes, not instance attributes.

- Decorator: @classmethod

- First parameter: cls (represents the class itself)

**Example:
class Student:**

    school_name = "ABC School"  # class attribute

    def __init__(self, name):
        self.name = name  # instance attribute

    @classmethod
    def change_school(cls, new_name):
        cls.school_name = new_name

    @classmethod
    def show_school(cls):
        print(f"School: {cls.school_name}")


# Using class method

- Student.show_school()  # School: ABC School

# Change class attribute using class method

- Student.change_school("XYZ School")
- Student.show_school()  # School: XYZ School


**Key Points:**

- Works with class-level data (cls.school_name).

- Can be called using the class or object.

Cannot access instance attributes directly (self.name).

**2. Static Methods***

- A static method does not access the instance (self) or class (cls).
It behaves like a regular function, but logically belongs to the class.

- Decorator: @staticmethod

- No default parameter like self or cls.

**Example:
class Calculator:**
    
    @staticmethod
    def add(a, b):
        return a + b
    
    @staticmethod
    def multiply(a, b):
        return a * b

# Calling static methods

- print(Calculator.add(5, 3))       # 8
- print(Calculator.multiply(5, 3))  # 15

**Key Points:**

- Does not need object or class reference.

- Cannot access instance attributes or class attributes.

- Mainly used for utility functions related to the class.

# **13.What is a property decorator in Python ?**


-  The @property decorator allows you to access a method like an attribute.
It is used to get, set, or delete an attribute in a controlled way while keeping encapsulation.

**Example:**

    class Student:
    def __init__(self, name):
        self._name = name  # private attribute

    @property
    def name(self):
        return self._name  # getter

    @name.setter
    def name(self, value):
        self._name = value  # setter

     s = Student("Alice")
      print(s.name)    # Alice  (calls getter)
      s.name = "Bob"   # calls setter
       print(s.name)    # Bob


**Explanation:**

- @property → makes a method behave like a read-only attribute (getter).

- @name.setter → allows controlled modification of the attribute.

# **17.What is the difference between a class variable and an instance variable?**

 **Class Variable vs Instance Variable**

**Class Variable:**

- Shared by all objects of a class.

- Declared inside class but outside methods.

- Can be accessed by ClassName.variable or self.variable.

- Used to store common data for all objects.

**Instance Variable:**

- Unique for each object of a class.

- Declared inside constructor __init__().

- Accessed using self.variable only.

- Used to store individual data for each object.

**Example:**

    class Student:
    school_name = "ABC School"  # class variable

    def __init__(self, name):
        self.name = name        # instance variable

    s1 = Student("Alice")
    s2 = Student("Bob")

    print(s1.name, s1.school_name)  # Alice ABC School
    print(s2.name, s2.school_name)  # Bob ABC School

    Student.school_name = "XYZ School"
    print(s1.school_name, s2.school_name)  # XYZ School XYZ School


**Summary:**

- Class variable → shared by all objects

- Instance variable → unique to each object

# **18.What is multiple inheritance in Python ?**


-  Multiple inheritance occurs when a child class inherits from more than one parent class.

**Purpose:**

- Allows a child class to reuse attributes and methods from multiple parent classes.

**Syntax:**

    class Parent1:
    def func1(self):
        print("Function of Parent1")

    class Parent2:
    def func2(self):
        print("Function of Parent2")

    class Child(Parent1, Parent2):  # inherits from both Parent1 and Parent2
    def func3(self):
        print("Function of Child")

    c = Child()
    c.func1()  # Function of Parent1
    c.func2()  # Function of Parent2
    c.func3()  # Function of Child


**Key Points:**

- Child class can access all methods from both parents.

- Python resolves conflicts using MRO (Method Resolution Order).

# **20.What is the significance of the ‘super()’ function in Python ?**

-  super() is a built-in function used to call a method from the parent (base) class.

**Purpose:**

- To avoid directly naming the parent class.

- Ensures proper method resolution, especially in multiple inheritance.

- Helps in code reuse by using the parent’s implementation.

**Example:**

    class Parent:
    def greet(self):
        print("Hello from Parent")

    class Child(Parent):
    def greet(self):
        super().greet()  # call parent method
        print("Hello from Child")

    c = Child()
    c.greet()


    Output:

    Hello from Parent
    Hello from Child


**Key Points:**

- Commonly used in inheritance.

- Helps in extending or customizing parent methods without rewriting them.

#**.22What is the difference between @staticmethod and @classmethod in Python?**


-   @staticmethod → Method that does not access instance (self) or class (cls).

- @classmethod → Method that accesses the class (cls), not the instance.

**First Parameter:**

- @staticmethod → No default parameter.

- @classmethod → cls, which represents the class.

**Access:**

- @staticmethod → Cannot access instance or class attributes.

- @classmethod → Can access class attributes but not instance attributes.

**Use Case:**

- @staticmethod → Used for utility functions related to the class.

- @classmethod → Used for modifying class-level data or factory methods.

**Example:**

    class Student:
    school_name = "ABC School"

    @staticmethod
    def greet():
        print("Welcome!")

    @classmethod
    def change_school(cls, name):
        cls.school_name = name

    Student.greet()              # Welcome!
    Student.change_school("XYZ") # Changes class attribute
    print(Student.school_name)   # XYZ School


**Summary:**

- @staticmethod → No access to class/instance, utility function.

- @classmethod → Access to class, modify class-level data.

# **23.How does polymorphism work in Python with inheritance ?**

**Key Points:**

- Achieved through method overriding in inheritance.

- Allows flexible and reusable code.

- Python’s duck typing also supports polymorphism without strict inheritance.

**Summary:
In Python, polymorphism with inheritance lets child classes override parent methods, so the same method call produces different behavior for different objects.**


# **24.What is method chaining in Python OOP?**

-   Method chaining is when multiple methods are called in a single statement on the same object.

- Achieved by having each method return self.

**Purpose:**

- Makes code compact and readable.

- Allows performing multiple operations on the same object in one line.

**Example:**

    class Student:
    def __init__(self, name):
        self.name = name

    def set_name(self, name):
        self.name = name
        return self  # return self for chaining

    def greet(self):
        print(f"Hello, {self.name}")
        return self  # return self for chaining

    s = Student("Alice")
    s.set_name("Bob").greet()  # Hello, Bob


**Explanation:**

- set_name() returns the object itself (self).

- greet() is called immediately after on the same object.

**Key Points:**

- Common in fluent interfaces.

- Each method must return self to continue chaining.

```python
"""1. Create a parent class Animal with a method speak() that prints a generic message. Create a child class Dog
that overrides the speak() method to print "Bark!"."""
# Parent class
class Animal:
    def speak(self):
        print("This animal makes a sound")

# Child class
class Dog(Animal):
    def speak(self):
        print("Bark!")

# Creating objects
a = Animal()
d = Dog()

# Calling methods
a.speak()  # This animal makes a sound
d.speak()  # Bark!
```

```python
"""2. Write a program to create an abstract class Shape with a method area(). Derive classes Circle and Rectangle
from it and implement the area() method in both."""

from abc import ABC, abstractmethod
import math

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

c = Circle(5)
r = Rectangle(4, 6)
print("Circle area:", c.area())
print("Rectangle area:", r.area())

```

```python
"""3. Implement a multi-level inheritance scenario where a class Vehicle has an attribute type. Derive a class Car
and further derive a class ElectricCar that adds a battery attribute."""
class Vehicle:
    def __init__(self, type):
        self.type = type

class Car(Vehicle):
    def __init__(self, type, brand):
        super().__init__(type)
        self.brand = brand

class ElectricCar(Car):
    def __init__(self, type, brand, battery):
        super().__init__(type, brand)
        self.battery = battery

ecar = ElectricCar("Car", "Tesla", "100 kWh")
print(ecar.type, ecar.brand, ecar.battery)
```

```python
"""5. Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes
balance and methods to deposit, withdraw, and check balance."""
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance  # private attribute

    def deposit(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
        else:
            print("Insufficient balance")

    def check_balance(self):
        return self.__balance

acc = BankAccount(1000)
acc.deposit(500)
acc.withdraw(300)
print("Balance:", acc.check_balance())
```

```python
"""7. Create a class MathOperations with a class method add_numbers() to add two numbers and a static
method subtract_numbers() to subtract two numbers."""
class MathOperations:
    @classmethod
    def add_numbers(cls, a, b):
        return a + b

    @staticmethod
    def subtract_numbers(a, b):
        return a - b

print(MathOperations.add_numbers(10, 5))     # 15
print(MathOperations.subtract_numbers(10, 5)) # 5
```

```python
"""10. Demonstrate operator overloading by creating a class Vector and overriding the add method to add two
vectors."""
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"({self.x}, {self.y})"

v1 = Vector(2, 3)
v2 = Vector(4, 5)
v3 = v1 + v2
print(v3)  # (6, 8)
```

```python
"""14. Create a class Employee with a method calculate_salary() that computes the salary based on hours worked
and hourly rate. Create a derived class Manager that adds a bonus to the salary."""
class Employee:
    def __init__(self, name, hours, rate):
        self.name = name
        self.hours = hours
        self.rate = rate

    def calculate_salary(self):
        return self.hours * self.rate

class Manager(Employee):
    def __init__(self, name, hours, rate, bonus):
        super().__init__(name, hours, rate)
        self.bonus = bonus

    def calculate_salary(self):
        return super().calculate_salary() + self.bonus

e = Employee("Alice", 40, 200)
m = Manager("Bob", 40, 200, 5000)

print("Employee Salary:", e.calculate_salary())
print("Manager Salary:", m.calculate_salary())
```

```python
"""16. Create a class Animal with an abstract method sound(). Create two derived classes Cow and Sheep that
implement the sound() method."""
from abc import ABC, abstractmethod
class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

# Derived class Cow
class Cow(Animal):
    def sound(self):
        print("Moo")

# Derived class Sheep
class Sheep(Animal):
    def sound(self):
        print("Baa")

# Create objects of derived classes only
animals = [Cow(), Sheep()]

# Call the sound method
for a in animals:
    a.sound()

```

```python
"""17. Create a class Book with attributes title, author, and year_published. Add a method get_book_info() that
returns a formatted string with the book's details."""

class Book:
    def __init__(self, title, author, year_published):
        self.title = title
        self.author = author
        self.year_published = year_published

    def get_book_info(self):
        return f"'{self.title}' by {self.author}, published in {self.year_published}"

b = Book("1984", "George Orwell", 1949)
print(b.get_book_info())
```
